```python
import numpy as np
import sys, os, git
from matplotlib import pyplot as plt
```

```python
current_path = os.getcwd()
git_repo = git.Repo(current_path, search_parent_directories=True)
git_path = git_repo.git.rev_parse("--show-toplevel")
sys.path.append(git_path + "/pyplot-perso/")
from functions import complete_panel, save_figure, set_boundaries, \
                      add_subplotlabels, set_boundaries, prepare_figure
from color_series1 import colors
path_figures = current_path[len(git_path):] + '/'
```

```python
filename = "potential-energy"
for mode, mygray in zip(['light', 'dark'], [colors["mylightgray"], colors["mydarkgray"]]):    
    fig = prepare_figure(mode, transparency = True, desired_figsize=(18,14))
    ax, n, l_tot, c_tot = [], 0, 3, 3
    n += 1
    ax.append(plt.subplot(l_tot, c_tot, n))
    t, energy = np.loadtxt("energy-em.xvg", skiprows=24).T
    ax[-1].plot(t/1000, energy/1000, "o", color=colors["mycyan"],
                markersize = 12, linewidth=4)
    complete_panel(ax[-1], r'$N_\textrm{steps} ~ (\times 1000)$',
                   r'$E_\mathrm{p} ~ \mathrm{(MJ/mol)}$',
                   xpad=10, legend=False, axis_color=mygray)
    set_boundaries(plt, x_boundaries=(-0.1, 5), y_boundaries=(-30, 0),
                   x_ticks=np.arange(0, 5.1, 1), y_ticks=np.arange(-30, 1, 10))
    
    n += 1
    ax.append(plt.subplot(l_tot, c_tot, n))
    t, energy = np.loadtxt("energy-nvt.xvg", skiprows=24).T
    ax[-1].plot(t, energy/1000, "o", color=colors["mycyan"],
                markersize = 12, linewidth=4)
    complete_panel(ax[-1], r'$t (\textrm{ps})$', None,
                   xpad=10, legend=False, axis_color=mygray)
    set_boundaries(plt, x_boundaries=(-1, 20), y_boundaries=(-30, 0),
                   x_ticks=np.arange(0, 20.1, 5), y_ticks=np.arange(-30, 1, 10))

    n += 1
    ax.append(plt.subplot(l_tot, c_tot, n))
    t, energy = np.loadtxt("energy-npt.xvg", skiprows=24).T
    ax[-1].plot(t, energy/1000, "o", color=colors["mycyan"],
                markersize = 12, linewidth=4)
    complete_panel(ax[-1], r'$t (\textrm{ps})$', None,
                   xpad=10, legend=False, axis_color=mygray)
    set_boundaries(plt, x_boundaries=(-1, 20), y_boundaries=(-30, 0),
                   x_ticks=np.arange(0, 20.1, 5), y_ticks=np.arange(-30, 1, 10))

    add_subplotlabels(fig, ax, [r"$\textrm{a}$", r"$\textrm{b}$", r"$\textrm{c}$"], color=mygray)
    save_figure(plt, fig, mode, git_path, path_figures, filename)
```
