```python
import copy
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.ticker import AutoMinorLocator
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Palatino"],
})
```

```python
fontsize = 20
font = {'family': 'sans', 'color':  'black', 'weight': 'normal', 'size': fontsize}
myblue = [0/ 255, 150/255, 177/ 255]
lightgray = [0.1, 0.1, 0.1]
darkgray = [0.9, 0.9, 0.9]
```

```python
short = np.loadtxt('short.xvg')
ys, xs = np.histogram(short.T[1], bins=50)
long = np.loadtxt('long.xvg')
yl, xl = np.histogram(long.T[1], bins=50)
```

```python
for mode, mygray in zip(['light', 'dark'], [lightgray, darkgray]):

    fig = plt.figure(figsize=(30, 4))
    ax1 = fig.add_subplot(131)
    plt.plot(xs[:-1], ys/np.sum(ys), linewidth = 2, label=r'50 ps', color=myblue) 
    plt.plot(xl[:-1], yl/np.sum(yl), linewidth = 2, label=r'1.5 ns', color=mygray) 
    #plt.plot(density_ethanol.T[0]-3.1, density_ethanol.T[1], linewidth = 2, color=myblue)
    #plt.plot(density_water.T[0]-3.1, density_water.T[1], linewidth = 2, color=mygray)

    divider = make_axes_locatable(ax1)
    ax1.set_xlabel('Distance (nm)', fontdict=font)
    ax1.set_ylabel('Probability', fontdict=font)
    plt.xticks(fontsize=fontsize)
    plt.yticks(fontsize=fontsize)
    #plt.ylim(5, 65)
    ax1.minorticks_on()
    ax1.tick_params('both', length=10, width=2, which='major', direction='in')
    ax1.tick_params('both', length=6, width=1.4, which='minor', direction='in')
    ax1.xaxis.set_ticks_position('both')
    ax1.yaxis.set_ticks_position('both')
    ax1.spines["top"].set_linewidth(2)
    ax1.spines["bottom"].set_linewidth(2)
    ax1.spines["left"].set_linewidth(2)
    ax1.spines["right"].set_linewidth(2)
    #ax1.set_xticks([0, 5, 10, 15, 20])
    #ax1.set_yticks([0, 0.4, 0.8, 1.2, 1.6])
    minor_locator_y = AutoMinorLocator(2)
    ax1.yaxis.set_minor_locator(minor_locator_y)
    minor_locator_x = AutoMinorLocator(2)
    ax1.xaxis.set_minor_locator(minor_locator_x)
    ax1.xaxis.label.set_color(mygray)
    ax1.yaxis.label.set_color(mygray)
    ax1.tick_params(axis='x', colors=mygray)
    ax1.tick_params(axis='y', colors=mygray)
    ax1.spines['left'].set_color(mygray)
    ax1.spines['top'].set_color(mygray)
    ax1.spines['bottom'].set_color(mygray)
    ax1.spines['right'].set_color(mygray)
    ax1.tick_params(axis='y', which='both', colors=mygray)
    ax1.tick_params(axis='x', which='both', colors=mygray)

    legend = ax1.legend(frameon=False, fontsize=fontsize,
            handlelength = 0.8, borderpad = 0.3, 
            labelspacing=0.3) 
    plt.setp(legend.get_texts(), color=mygray)

    fig.tight_layout()
    if mode == 'light':
        pass
        #plt.savefig('temperature.png', bbox_inches = 'tight', pad_inches = 0.057, transparent=True)
        plt.savefig('../../../../../docs/doc-sphinx/source/tutorials/figures/ethanoladsorption/probability-light.png', bbox_inches = 'tight', pad_inches = 0.057, transparent=True)
    else:
        pass
        #plt.savefig('temperature-dark.png', bbox_inches = 'tight', pad_inches = 0.057, transparent=True)
        plt.savefig('../../../../../docs/doc-sphinx/source/tutorials/figures/ethanoladsorption/probability-dark.png', bbox_inches = 'tight', pad_inches = 0.057, transparent=True)        
    plt.show()
```
